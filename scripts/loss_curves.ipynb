# Visualizing the Results 

```python
#plotting bar graph with bins of different colors
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
SMALL_SIZE = 16
MEDIUM_SIZE = 9
BIGGER_SIZE = 14

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
epochs=range(0,6500,500)
```

### Creating dataframes for Plotting

```python
focal_loss = pd.DataFrame(focal_loss, index =epochs,columns =['focal_loss']) 
focal_symbol_error_rate = pd.DataFrame(focal_symbol_error_rate, index =epochs,columns =['focal_symbol_error_rate']) 
focal_sequence_error_rate = pd.DataFrame(focal_sequence_error_rate, index =epochs,columns =['focal_sequence_error_rate']) 
traditional_loss = pd.DataFrame(traditional_loss, index =epochs,columns =['traditional_loss']) 
traditional_symbol_error_rate = pd.DataFrame(traditional_symbol_error_rate, index =epochs,columns =['traditional_symbol_error_rate']) 
traditional_sequence_error_rate = pd.DataFrame(traditional_sequence_error_rate, index =epochs,columns =['traditional_sequence_error_rate']) 
```

### Converting them into float

```python
focal_loss=focal_loss.astype(float)
focal_symbol_error_rate=focal_symbol_error_rate.astype(float)
focal_sequence_error_rate=focal_sequence_error_rate.astype(float)
traditional_loss=traditional_loss.astype(float)
traditional_symbol_error_rate=traditional_symbol_error_rate.astype(float)
traditional_sequence_error_rate=traditional_sequence_error_rate.astype(float)
```

```python
f, ax = plt.subplots(figsize=(12,4)) # set the size that you'd like (width, height)

plt.title('Training Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')

plt.plot(traditional_loss,label='Traditional Loss')
plt.plot(focal_loss,label='Focal Loss (Ours)')

plt.yscale('log')

plt.legend()

plt.savefig('TrainingLoss.png')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(12,4)) # set the size that you'd like (width, height)

plt.title("Comparing Symbol Error Rate")
plt.ylabel('Symbol Error Rate(%)')
plt.xlabel('Epochs')

plt.plot(traditional_symbol_error_rate,label='Traditional Model')
plt.plot(focal_symbol_error_rate,label='Focal Model (Ours)')

plt.yscale('log')

plt.legend()

plt.savefig('ComparingSymbolErrorRate.png')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(12,4)) # set the size that you'd like (width, height)

plt.title("Comparing Sequence Error Rate")
plt.ylabel('Sequnce Error Rate')
plt.xlabel('Epochs')


plt.plot(traditional_sequence_error_rate,label='Traditional Model')
plt.plot(focal_sequence_error_rate,label='Focal Model (Ours)')

plt.yscale('log')

plt.legend()

plt.savefig('ComparingSequenceErrorRate.png')
plt.show()
```
